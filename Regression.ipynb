```python
import pandas as pd
```

```python
df = pd.read_csv('dataset_clean.csv')
```

### OLS on Lag1

```python
import statsmodels.api as sms

X1 = sms.add_constant(df.lag1)
y1 = df.price

model1 = sms.OLS(y1, X1)
results1 = model1.fit()
results1.summary()
```
